# Escalares, vectores, matrices y tensores 
En el siguiente notebook se usará la librería Torch para declarar objetos matemáticos tales como escales, vectores, matrices y tensores (que al final de cuentas todos pueden ser considerados como tensores). Está inspirado en una clase en youtube que puedes encontrar en el siguiente link: https://youtu.be/fweUebNthDE?si=Z18wOtufg5AqLR2v

```python
# Importando las librerias necesarias
import torch 
import matplotlib.pyplot as plt
```

```python
# Definiendo un escalar 
escalar = torch.tensor(1)
type_escalar = type(escalar)
print('Este es un escalar declarado en torch: ', escalar)
print('Este es el tipo de objeto al que pertenece dicho escalar: ', type_escalar)
```

```python
# Trabajando con vectores 
# Se define al vector
vector = torch.tensor([2.0,2.0])
print('Este es un vector declarado con torch: ', vector)

# Operando estos objetos (suma (regla del paralelogramo), norma y producto punto) 
v1 = torch.tensor([4.0,1.0])
v2 = torch.tensor([2.0,4.0])
suma_v = v1 + v2
print('Este es el resultado de la suma de los vectores v1 y v2', suma_v)

# Calculando normas 
# Norma eucluidianan L2 
norma_l2 = torch.norm(v1, p=2)
print('Esta es la norma del v1' , norma_l2)

#Producto punto 
dot = torch.dot(v1,v2)
print(f"Producto punto de {v1.numpy()} y {v2.numpy()}: {dot.item()}")

# Visualizacion los vectores y de su suma 
plt.figure(figsize=(4, 4))
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='purple', label='Vector 1')
plt.quiver(v1[0], v1[1], v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='g', label='Vector 2 (desde v1)')
plt.quiver(0,0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='mediumseagreen', label='Vector 2 (desde el origen)')
plt.quiver(0, 0, suma_v[0], suma_v[1], angles='xy', scale_units='xy', scale=1, color='pink', label='Vector 1 + Vector 2')
plt.xlim(0, 7)
plt.ylim(0, 7)
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Suma de Vectores')
plt.legend()
plt.grid()
plt.show()
```

```python
# Trabajando con matrices 
m1 = torch.tensor([[1,2],[2,2]])
m2 = torch.tensor([[3,3],[4,9]])
#Suma
suma_m = m1 + m2
print(f"Suma de las matrices m1 + m2:\n{suma_m}")
#Multiplicacion
C = torch.matmul(m1, m2)
print(f"Producto matricial m1 x m2:\n{C}")

# Trabajando con tensores
tensor = torch.rand(2,2,3)
print(f"Ejemplo de un tensor 2x2x3 (son dos matrices con dos filas y tres columnas cada una):\n{tensor}")
print(tensor[1,1,0])

# Extraer un subtensor 
sub_tensor = tensor[0,:,:] # Se extrae la primera matriz de 2x3 de mi tensor original, con esto se le baja una dimension al tensor 
print('Este es mi subtensor: ', sub_tensor)
```

Se tiene un conjunto de 64 imágenes de 28 x 28 pixeles

```python
# Supongase una capa con 784 entradas y 128 neuronas
# Se definen los pesos y sesgos de la siguiente manera 
pesos = torch.rand(784, 128)
sesgos = torch.rand(128)
print(f"Dimensiones de los pesos: {pesos.shape}")
print(f"Dimensiones de los sesgos: {sesgos.shape}")

# Entrada de la red (por ejemplo, una imagen aplanada de 28x28 píxeles)
entrada = torch.rand(1, 784)  # Batch size de 1

# Operación de la capa (entrada x pesos + sesgos)
salida = torch.matmul(entrada, pesos) + sesgos
print(f"Salida de la capa: {salida.shape}")
```
